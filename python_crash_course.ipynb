### Dictionary in python

Defination: \
A dictionary in Python is a collection of data values, used to store information like a map

Unlike other Python data types, a dictionary holds key-value pairs, where each key maps to a value, allowing efficient access and retrieval of data.

```python
# initialize empty dictionary
d = {}

d = {1: 'Hello', 2: 'guys', 3: 'How are you?'}
print(d)

# creating dictionary using dict() constructor
d1 = dict({1: 'Hello', 2: 'guys', 3: 'How are you?'})
print(d1)

d = {1: 'hello', 'name': 'guys', 3: 'How are you'}

# Accessing an element using key
print(d['name'])

# Accessing a element using get
print(d.get(3))
print(d.get(4, 0))  #if the key is not in the dictionary then return None by default here we make it to return "0"

#Dictionary comprehension
divisible_2 = {x: x%2==0 for x in range(10)}
print(divisible_2)

# Keys as numbers, values as their squares
squares = {x: x**2 for x in range(10)}
print(squares)
```

### Functions in python

defination: \
Python Functions are a block of statements that does a specific task. \
It is used to make code or blocks of code reusable


image.png

```python
#simple function:
def fun():
    print("Welcome to python crash couse")

fun()

#passing arguments and using parameters in function
def evenOdd(x):           #"x" is a parameter here
    if (x % 2 == 0):
        return "Even"
    else:
        return "Odd"

print(evenOdd(16))    #16 is an argument that is being passed
print(evenOdd(7))

#default arguments
def myFun(x, y=50):
    print("x: ", x)
    print("y: ", y)

myFun(10)

#position of the arguments passed is important
def nameAge(name, age):
    print("Hi, I am", name)
    print("My age is ", age)

print("Case-1:")
nameAge("Olivia", 27)

print("\nCase-2:")
nameAge(27, "Olivia")

#Arbitory Arguments or variable number of arguments
#Defination:
# In Python, arbitrary arguments allow a function to accept a variable number of inputs. This is done using two special symbols:
# -- *args: collects extra positional (non-keyword) arguments as a tuple.
# -- **kwargs: collects extra keyword arguments as a dictionary.

def myFun(*args, **kwargs):
    print("Non-Keyword Arguments (*args):")
    for arg in args:
        print(arg)

    print("\nKeyword Arguments (**kwargs):")
    for key, value in kwargs.items():
        print(f"{key} == {value}")

myFun('Hey', 'Welcome', first='python', mid='crash', last='course')

# Nested functions:
def f1():
    s = 'python is fun'
    def f2():
        print(s)

    f2()
f1()


#Anonymous Functions or single liner function
def c1(x): return x*x*x
c2 = lambda x : x*x*x

print(c1(7))
print(c2(7))

#using return in functions
def sq_value(num):
    """This function returns the square
    value of the entered number"""
    return num**2

print(sq_value(2))
print(sq_value(-4))

#Recursion in python:
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(4))
```

Python Exception Handling allows a program to gracefully handle unexpected events (like invalid input or missing files) without crashing

### basic structure of OOPS:

Classes: They are the blueprints of the OOPs, they tell how the functions and methods looks \
Attributes are the variables that belong to a class. \
Attributes are always public and can be accessed using the dot (.) operator.

What is __init__ do? \
It is called as the dunder method or the magic method that is used to construct class attribute. It is used to initialize the objects of the class. By default Python does this.

What is the importance of self? \
It is the way an object(variable) talks about itself. Without self, a class wouldn't know which specific object's data it should be changing.

```python
# Basic oops program

class Dog:
    species = "Canine"  # Class attribute

    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age  # Instance attribute

# Creating an object of the Dog class
dog1 = Dog("Buddy", 3)

print(dog1.name)
print(dog1.species)
```

### Inheritance

Defination: \
Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a class (called a child or derived class) to inherit attributes and methods from another class (called a parent or base class)

```python
# Parent Class: Animal
class Animal:
    def __init__(self, name):
        self.name = name

    def info(self):
        print("Animal name:", self.name)

# Child Class: Dog
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)   # Call parent constructor
        self.breed = breed

    def details(self):
        print(self.name, "is a", self.breed)

d = Dog("Buddy", "Golden Retriever")
d.info()      # Parent method
d.details()   # Child method
```

### Polymorphism

Defination: \
Polymorphism means "many forms". It refers to the ability of an entity (like a function or object) to perform different actions based on the context.

```python
#Method Overriding

class Animal:
    def sound(self):
        return "Some generic sound"

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"

# Polymorphic behavior
animals = [Dog(), Cat(), Animal()]
for animal in animals:
    print(animal.sound())
```

```python
## Method overloading:

class Calculator:
    def calculate_area(self, length, width=None):
        if width is not None:
            #it's a Rectangle
            return length * width
        else:
            #it's a Square
            return length * length

# creating the object
calc = Calculator()

# Scenario 1: Square (One argument)
print(f"Square Area: {calc.calculate_area(5)}")

# Scenario 2: Rectangle (Two arguments)
print(f"Rectangle Area: {calc.calculate_area(5, 10)}")
```
